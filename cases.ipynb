```python
repository = 'https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/'
url = 'dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv'
pr_url = 'dati-province/dpc-covid19-ita-province.csv'
r_url = 'dati-regioni/dpc-covid19-ita-regioni.csv'
```

```python
df = pd.read_csv(repository+url)
pr_df = pd.read_csv(repository+pr_url)
r_df = pd.read_csv(repository+r_url)
```

```python
abr_df = r_df[r_df['denominazione_regione'] == 'Abruzzo']
```

```python
c, t, d = df[['totale_casi', 'terapia_intensiva', 'deceduti']].values.T
abr_c, abr_t, abr_d = abr_df[['totale_casi', 'terapia_intensiva', 'deceduti']].values.T
```

```python
plt.plot((abr_t/abr_c)[-100:], label='abruzzo')
plt.plot((t/c)[-100:], label='nazionale')
plt.legend()
```

```python
plt.plot((abr_d/abr_c)[-100:], label='abruzzo')
plt.plot((d/c)[-100:], label='nazionale')
plt.legend()
```
